## 2 Рівномірний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html).

```python
from scipy.stats import uniform
```

```python
a = 0
b = 1
```

Параметри $loc, scale$ відповідають рівномірному розподілу на відрізку $[loc, loc + scale]$.

```python
loc = a
scale = b - loc
```

### 2.1 Щільність

```python
x1 = np.arange(-1, 2, 0.1)
```

Виведіть значення функції щільності рівномірного розподілу з параметрами $loc, scale$ в точках вектора $x1$.

```python
y1 = uniform.pdf(x1, loc=loc, scale=scale)
print(y1)
```

Зобразіть графічно значення функції щільності рівномірного розподілу з параметрами $loc, scale$ в точках вектора $x2$.

```python
x2 = np.arange(-1, 2, 0.01)
```

```python
y2 = uniform.pdf(x2, loc=loc, scale=scale)
plt.plot(x2, y2)
plt.show()
```

### 2.2 Функція розподілу

Виведіть значення функції розподілу рівномірного розподілу з параметрами $loc, scale$ в точках вектора $x1$.

```python
y1 = uniform.cdf(x1, loc=loc, scale=scale)
print(y1)
```

Зобразіть графічно значення функції розподілу рівномірного розподілу з параметрами $loc, scale$ в точках вектора $x2$.

```python
y2 = uniform.cdf(x2, loc=loc, scale=scale)
plt.plot(x2, y2)
plt.show()
```

### 2.3 Випадкові величини

Згенеруйте $100$ випадкових величин з рівномірного розподілу з параметрами $loc, scale$.

```python
unif_rvs = uniform.rvs(loc=loc, scale=scale, size=100)
print(unif_rvs)
```

Зобразіть гістограму $10000$ випадкових величин з рівномірного розподілу з параметрами $loc, scale$.

```python
unif_rvs = uniform.rvs(loc=loc, scale=scale, size=10000)
plt.hist(unif_rvs)
plt.show()
```

Зобразіть графік функції щільності та нормовану гістограму $100000$ випадкових величин з рівномірного розподілу з параметрами $loc, scale$.

## 5 Гамма розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html).

```python
from scipy.stats import gamma
```

```python
alpha = 2
```

Параметри $loc, scale$ відповідають гамма розподілу зі щільністю $$f(x) = \frac{1}{\Gamma(\alpha) \, scale} \left( \frac{x - loc}{scale} \right)^{\alpha - 1} \exp \left\{ -\frac{x - loc}{scale} \right\}.$$

```python
loc = 0
scale = 1
```

### 5.1 Щільність

```python
x1 = np.arange(-1, 2.1, 0.1)
```

Виведіть значення функції щільності гамма розподілу з параметрами $alpha, loc, scale$ в точках вектора $x1$.

```python
# Місце для Вашого коду
y1 = gamma.pdf(x1, alpha, loc=loc, scale=scale)
print(y1)
```

Зобразіть графічно значення функції щільності гамма розподілу з параметрами $alpha, loc, scale$ в точках вектора $x2$.

```python
x2 = np.arange(-1, 7.01, 0.01)
```

```python
# Місце для Вашого коду
y2 = gamma.pdf(x2, alpha, loc=loc, scale=scale)
plt.plot(x2, y2)
plt.show()

```

### 5.2 Функція розподілу

Виведіть значення функції розподілу гамма розподілу з параметрами $alpha, loc, scale$ в точках вектора $x1$.

```python
# Місце для Вашого коду
y1 = gamma.cdf(x1, alpha, loc=loc, scale=scale)
print(y1)

```

Зобразіть графічно значення функції розподілу гамма розподілу з параметрами $alpha, loc, scale$ в точках вектора $x2$.

```python
# Місце для Вашого коду
fig, axs = plt.subplots(1,2, figsize=(20,5))


print(fig)
y2 = gamma.pdf(x2, alpha, loc=loc, scale=scale)
y2_ = gamma.cdf(x2, alpha, loc=loc, scale=scale)
axs[0].plot(x2, y2_, 'r--')
axs[1].plot(x2, y2_, 'r--', x2, y2, 'b--' )
axs[1].set_title('cdf+pdf')
```

## 6 Апроксимація геометричного розподілу експоненційним розподілом

Покажіть схожість геометричного розподілу та експоненційного розподілу з відповідними параметрами:
- $p = 0.1$
- $p = 0.2$
- $p = 0.3$
- $p = 0.4$
- $p = 0.5$
- $p = 0.6$
- $p = 0.7$
- $p = 0.8$
- $p = 0.9$

Зобразіть графіки їх функцій щільності (обидві функції на одному графіку) та функцій розподілу (обидві функції на другому графіку) на двох сабплотах в одному рядку.

**Зауваження 1.** Парамери розподілів слід брати однаковими ($p = \lambda$).

**Зауваження 2.** Для порівняння розподілів треба "зсунути" геометричний розподіл на одиницю вліво, тобто щоб його значення починалися не з $1$, а з $0$.

```python
# Місце для Вашого коду
from scipy.stats import geom
for i in range (1, 10):
    p=i/10
    lmd=p
    loc = 0
    scale = 1 / lmd
    x2 = np.arange(-1, 50, 1)
    y2 = expon.pdf(x2, loc=loc, scale=scale)
    y2_ = gamma.cdf(x2, alpha, loc=loc, scale=scale)
    fig, axs = plt.subplots(1,2, figsize=(20,5))
    yg = geom.pmf(x2, p)
    yg_ = geom.cdf(x2, p)
    
    axs[0].plot(x2, y2, 'r--', x2-1, yg, 'b')
    axs[0].set_title('pdf')
    axs[1].plot(x2, y2_, 'r--', x2-1, yg_, 'b')
    axs[1].set_title('cdf')
```
